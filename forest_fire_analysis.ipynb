```r
library(tidyverse)
#install.packages('ggthemes')
library(ggthemes)
```

```r
url = 'https://raw.githubusercontent.com/patcarrasco/forestfiresimulator/master/forest_fire_data.csv'

fire_data  <- read.csv(url)
```

```r
head(fire_data)
```

```r
fire_data$prop_alive = as.character(fire_data$prop_alive)
fire_data$time = as.double(fire_data$time)
```

```r
ggplot(data = fire_data, mapping = aes(x = time, y = alive)) +
  geom_jitter(aes(color = prop_alive), size = 0.1) +
  theme_base() +
  guides(colour = guide_legend(override.aes = list(size=7))) +
  scale_colour_calc()
```

```r
ggplot(data = fire_data, mapping = aes(x = time, y = dead)) +
  geom_jitter(aes(color = prop_alive), size = 0.1) +
  theme_base() +
  guides(colour = guide_legend(override.aes = list(size=7))) +
  scale_colour_calc()
```

```r
ggplot(data = fire_data, mapping = aes(x = time, y = onfire)) +
  geom_jitter(aes(color = prop_alive), size = 0.1) +
  theme_base() +
  guides(colour = guide_legend(override.aes = list(size=7))) +
  scale_colour_calc()
```

```r
ggplot(data = fire_data, mapping = aes(x = time, y = onfire)) +
  geom_path(aes(color = prop_alive), size = 0.1, alpha = 1) +
  theme_base() +
  guides(colour = guide_legend(override.aes = list(size=7))) +
  scale_colour_calc()
```

```r
ggplot(data = fire_data, mapping = aes(x = prop_alive, y = time)) +
  geom_boxplot(aes(fill = prop_alive)) +
  theme_base()
```

```r
ggplot(data = fire_data, mapping = aes(x = time, fill = prop_alive)) +
  geom_histogram(position = 'dodge', binwidth = 20) +
  theme_base() +
  scale_colour_calc()
```
